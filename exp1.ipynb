### Imports

```python
import random, time 
import torch
import torch.nn
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
from torchvision.models import resnet18, ResNet18_Weights
```

### Modèles

```python
def resnet18_scratch(num_classes = 10) :
    model = resnet18(weights = None)
    # Premiere conv doit être modifier pour adapter à noir et blanc cad 1 canal
    model.conv1 = nn.Conv2d(1,64, kernel_size = 7, stride = 2, padding = 3, biais = False)
    # Adapter la tête aux classes de notre cas
    model.fc = nn.Linear(model.fc.in_features, num_classes)
    return model
```

```python
def resnet18_pretrained(num_classes = 10):
    model = resnet18(weigths = True)
    model.fc = nn.Linear(model.fc.in_fetures, num_classes)
    return model
```

#### Boucles entrainement / evaluation

```python
def train_1_epoch(model, loader, optimizer, criterion, device):
    model.train()
    total_loss, total_correct, total = 0.0,0,0
    for x,y in loader :
        x, y = x.to(device), y.to(device)
        optimizer.zero_grad(set_to_none= True)
        logits = model(x)
        loss = criterion(logits, y)
        loss.backward()
        optimizer.step()

        total_loss += loss.item() * x.size(0)
        total_correct+= (logits.argmax(1) == y).sum().item()
        total += x.size(0)
    return total_loss / total_correct / total
```

```python
def evaluate(model, loader, criterion, device):
    model.eval()
    total_loss, total_correct, total = 0.0,0,0
    for x, y in loader :
        x, y = x.to(device), y.to(device)
        logits = model(x)
        loss = criterion(logits,y)


        total_loss += loss.item() * x.size(0)
        total_correct+= (logits.argmax(1) == y).sum().item()
        total += x.size(0)
    return total_loss / total_correct / total
```
